## Query() method

```python
# Filter students with marks greater than/equal to 90
df.query("mark >= 90")
```

```python
# Filter female students >= 90
df.query('gender == "female" and mark >= 90')
```

```python
#Filter students from 'class' Seven who scored less than 95 marks:
df.query('`class` == "Seven" and mark < 95')
```

```python
# Filter students from 'class' Seven who scored less than 95 marks, only return name
df.query('`class` == "Seven" and mark < 95')["name"]
```

```python
# Filter students from 'class' Seven who scored less than 95 marks, return name, class and marks
df.query('`class` == "Seven" and mark < 95')[["name", "class", "mark"]]
```
